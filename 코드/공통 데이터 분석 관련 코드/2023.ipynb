# 혼자 여행하는 관광객 데이터 추출

```python
# 관광/휴양을 목적으로 한 혼여행객 데이터 추출

df_여행1차 = df[
    (df['D_TRA1_CASE'] == 1) &
    (df['D_TRA1_NUM'] == 1)
]

df_여행2차 = df[
    (df['D_TRA2_CASE'] == 1) &
    (df['D_TRA2_NUM'] == 1)
]

df_여행3차 = df[
    (df['D_TRA3_CASE'] == 1) &
    (df['D_TRA3_NUM'] == 1)
]

df_여행4차 = df[
    (df['D_TRA4_CASE'] == 1) &
    (df['D_TRA4_NUM'] == 1)
]

df_여행5차 = df[
    (df['D_TRA5_CASE'] == 1) &
    (df['D_TRA5_NUM'] == 1)
]

df_여행6차 = df[
    (df['D_TRA6_CASE'] == 1) &
    (df['D_TRA6_NUM'] == 1)
]

dfs = [df_여행1차, df_여행2차, df_여행3차, df_여행4차, df_여행5차, df_여행6차]

df_여행_전체 = pd.concat(dfs, ignore_index=True)
```

```python
df_여행_전체
```

```python
col_nums = ['D_TRA1_NUM', 'D_TRA2_NUM', 'D_TRA3_NUM', 'D_TRA4_NUM', 'D_TRA5_NUM', 'D_TRA6_NUM']
col_cases = ['D_TRA1_CASE', 'D_TRA2_CASE', 'D_TRA3_CASE', 'D_TRA4_CASE', 'D_TRA5_CASE', 'D_TRA6_CASE']

conditions = False
for case_col, num_col in zip(col_cases, col_nums):
    conditions |= (df[case_col] == 1) & (df[num_col] == 1)

travel_alone = df[conditions]
```

```python
travel_alone
```

```python
travel_alone.to_csv('travel_alone_2023.csv',index=False)
```

```python
# 여행 1차부터 6차까지 혼자 여행한 적이 1번이라도 있는 데이터 추출
conditions = (
    ((df['D_TRA1_CASE'] == 1) & (df['D_TRA1_NUM'] == 1)) |
    ((df['D_TRA2_CASE'] == 1) & (df['D_TRA2_NUM'] == 1)) |
    ((df['D_TRA3_CASE'] == 1) & (df['D_TRA3_NUM'] == 1)) |
    ((df['D_TRA4_CASE'] == 1) & (df['D_TRA4_NUM'] == 1)) |
    ((df['D_TRA5_CASE'] == 1) & (df['D_TRA5_NUM'] == 1)) |
    ((df['D_TRA6_CASE'] == 1) & (df['D_TRA6_NUM'] == 1))
)

col = ['D_TRA1_NUM', 'D_TRA2_NUM', 'D_TRA3_NUM', 'D_TRA4_NUM', 'D_TRA5_NUM', 'D_TRA6_NUM']

travel_alone = df[df[col].eq(1).any(axis=1)]
```

```python
cols = ['D_TRA1_CASE', 'D_TRA2_CASE', 'D_TRA3_CASE', 'D_TRA4_CASE', 'D_TRA5_CASE', 'D_TRA6_CASE']

# 값이 한번이라도 2인 행만 필터링: 관광/휴향 목적 여행인 사람 필터링
travel_heal = df[df[cols].eq(1).any(axis=1)]

col = ['D_TRA1_NUM', 'D_TRA2_NUM', 'D_TRA3_NUM', 'D_TRA4_NUM', 'D_TRA5_NUM', 'D_TRA6_NUM']

# 값이 한번이라도 1인 행만 필터링: 혼자 여행을 해 본 사람 필터링
travel_heal_1 = travel_heal[travel_heal[col].eq(1).any(axis=1)]

print(len(travel_heal_1))
```

```python
travel_heal_1.to_csv('travel_alone_2023_diff.csv',index=False)
```
