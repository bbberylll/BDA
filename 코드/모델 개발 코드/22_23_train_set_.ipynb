22년 데이터

```python
phases = [1, 2, 3, 4, 5, 6]
days = list(range(1, 18))

dfs = []

for i in phases:
    for j in days:
        col_name = f'D_TRA{i}_{j}_SPOT'
        if col_name in travel_22.columns:
            df_temp = travel_22[['ID']].copy()
            df_temp['차수'] = i
            df_temp['여행 인원'] = travel_22.get(f'D_TRA{i}_NUM')
            df_temp['지출 비용'] = travel_22.get(f'D_TRA{i}_COST')
            df_temp['방문 지역'] = travel_22.get(col_name)
            df_temp['여행 일수'] = travel_22.get(f'D_TRA{i}_S_Day')

            df_temp['철도비'] = travel_22.get('A9F_1')
            df_temp['도로비'] = travel_22.get('A9F_4')
            df_temp['식음료비'] = travel_22.get('A9E')
            df_temp['음식점비'] = travel_22.get('A9D')
            df_temp['쇼핑비'] = travel_22.get('A9H')
            df_temp['숙박비'] = travel_22.get('A9C')
            df_temp['교통비'] = travel_22.get('A9F')
            df_temp['여행 활동비'] = travel_22.get('A9G')
            dfs.append(df_temp)

long_df = pd.concat(dfs, ignore_index=True)
long_df = long_df.dropna(subset=['방문 지역'])

long_df['여행 인원'] = pd.to_numeric(long_df['여행 인원'], errors='coerce')
long_df['지출 비용'] = pd.to_numeric(long_df['지출 비용'], errors='coerce')
long_df['방문 지역'] = pd.to_numeric(long_df['방문 지역'], errors='coerce')
long_df['여행 일수'] = pd.to_numeric(long_df['여행 일수'], errors='coerce')
long_df['철도비'] = pd.to_numeric(long_df['철도비'], errors='coerce')
long_df['도로비'] = pd.to_numeric(long_df['도로비'], errors='coerce')
long_df['식음료비'] = pd.to_numeric(long_df['식음료비'], errors='coerce')
long_df['음식점비'] = pd.to_numeric(long_df['음식점비'], errors='coerce')
long_df['쇼핑비'] = pd.to_numeric(long_df['쇼핑비'], errors='coerce')
long_df['숙박비'] = pd.to_numeric(long_df['숙박비'], errors='coerce')
long_df['교통비'] = pd.to_numeric(long_df['교통비'], errors='coerce')


alone22_df = long_df[long_df['여행 인원'] == 1]
```

23년 데이터

```python
phases = [1, 2, 3, 4, 5, 6]
days = list(range(1, 18))

dfs = []

for i in phases:
    for j in days:
        col_name = f'D_TRA{i}_{j}_SPOT'
        if col_name in travel_23.columns:
            df_temp = travel_23[['ID']].copy()
            df_temp['차수'] = i
            df_temp['여행 인원'] = travel_23.get(f'D_TRA{i}_NUM')
            df_temp['지출 비용'] = travel_23.get(f'D_TRA{i}_COST')
            df_temp['방문 지역'] = travel_23.get(col_name)
            df_temp['여행 일수'] = travel_23.get(f'D_TRA{i}_S_Day')

            df_temp['철도비'] = travel_23.get('A9F_1')
            df_temp['도로비'] = travel_23.get('A9F_4')
            df_temp['식음료비'] = travel_23.get('A9E')
            df_temp['음식점비'] = travel_23.get('A9D')
            df_temp['쇼핑비'] = travel_23.get('A9H')
            df_temp['숙박비'] = travel_23.get('A9C')
            df_temp['교통비'] = travel_23.get('A9F')
            df_temp['여행 활동비'] = travel_23.get('A9G')

            dfs.append(df_temp)

long_df = pd.concat(dfs, ignore_index=True)
long_df = long_df.dropna(subset=['방문 지역'])

long_df['여행 인원'] = pd.to_numeric(long_df['여행 인원'], errors='coerce')
long_df['지출 비용'] = pd.to_numeric(long_df['지출 비용'], errors='coerce')
long_df['방문 지역'] = pd.to_numeric(long_df['방문 지역'], errors='coerce')
long_df['여행 일수'] = pd.to_numeric(long_df['여행 일수'], errors='coerce')
long_df['철도비'] = pd.to_numeric(long_df['철도비'], errors='coerce')
long_df['도로비'] = pd.to_numeric(long_df['도로비'], errors='coerce')
long_df['식음료비'] = pd.to_numeric(long_df['식음료비'], errors='coerce')
long_df['음식점비'] = pd.to_numeric(long_df['음식점비'], errors='coerce')
long_df['쇼핑비'] = pd.to_numeric(long_df['쇼핑비'], errors='coerce')
long_df['숙박비'] = pd.to_numeric(long_df['숙박비'], errors='coerce')
long_df['교통비'] = pd.to_numeric(long_df['교통비'], errors='coerce')

alone23_df = long_df[long_df['여행 인원'] == 1]
```

```python
grouped_22_23_df = pd.concat([alone22_df, alone23_df])

grouped_22_23_df = grouped_22_23_df.groupby('방문 지역').agg({
    '지출 비용': 'mean',
    # '지출 비용': 'sum',
    '철도비': 'mean',
    '도로비': 'mean',
    '식음료비': 'mean',
    '음식점비': 'mean',
    '쇼핑비': 'mean',
    '여행 일수': 'mean',
    '숙박비': 'mean',
    '교통비': 'mean',
    '여행 활동비': 'mean'
}).reset_index()

grouped_22_23_df['방문 지역'] = grouped_22_23_df['방문 지역'].astype(int)

print(grouped_22_23_df.head())
```
