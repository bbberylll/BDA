# 국가철도공단_철도역 정보_20240715

```python
import pandas as pd
data=pd.read_csv('/content/국가철도공단_철도역 정보_20240715.csv', encoding='euc-kr')
```

```python
data.columns
```

```python
data=data[['주소','역이름']]
data
```

```python
# 시도와 시군구 나누는 함수
def extract_sido(address):
    return address.split()[0] if isinstance(address, str) and len(address.split()) >= 1 else None

def extract_sigungu(address):
    return address.split()[1] if isinstance(address, str) and len(address.split()) >= 2 else None

# 새 컬럼 생성
data['시도'] = data['주소'].apply(extract_sido)
data['시군구'] = data['주소'].apply(extract_sigungu)

# 결과 확인
print(data[['주소', '시도', '시군구']].head())
```

```python
# 주소에서 시도/시군구 추출 함수
def split_address(address):
    parts = address.split()
    if len(parts) >= 2:
        return f"{parts[0]} {parts[1]}"
    else:
        return None

# 새로운 열 추가
data['시도_시군구'] = data['주소'].apply(split_address)

# 결과 확인
print(data[['주소', '시도_시군구']].head())
```

```python
data['시도'].value_counts()
```

```python
sido_mapping = {
    '전남': '전라남도',
    '서울시': '서울특별시',
    '경남': '경상남도'
}

# 시도 컬럼에 적용
data['시도'] = data['시도'].replace(sido_mapping)
```

```python
data['시도'].value_counts()
```

```python
data
```

```python
data['지역'] = data['시도'] + ' ' + data['시군구']
```

```python
data.drop('시도_시군구', axis=1, inplace=True)
```

```python
data
```

지역명코드 이용

```python
code=pd.read_excel('/content/code.xlsx')
```

```python
code['지역 코드'] = code['시/도 코드']*1000+ code['시/군/구 코드']
```

```python
code['지역'] = code['시/도명'] + ' ' + code['시/군/구명']
```

```python
code.drop(['시/도 코드', '시/도명', '시/군/구 코드', '시/군/구명'],axis=1, inplace=True)
```

시군구별 철도역 개수 계산

```python
# '지역' 컬럼 기준으로 병합
result= data.merge(code, on='지역', how='left')

# 결과 확인
print(result.head())
```

```python
result.drop(['주소','시도','시군구'], axis=1, inplace=True)
```

```python
result[result['지역 코드'].isna()]
```

```python
result
```

```python
train=result.groupby('지역 코드').size().reset_index(name='철도역개수')
```

# 대중교통 통행량_2024

```python
import pandas as pd
# 엑셀 파일 불러오기
file_path = "이용량 지표(수단통행량)_2024 (1).xlsx"
df = pd.read_excel(file_path, sheet_name='Main View')

# 병합 셀 문제 해결: 위쪽 값으로 채우기
df = df.ffill()

# 공항버스 제외 & 잘못된 시도/시군구 제거
df_filtered = df[(df['교통수단'] != '공항버스') & (df['교통수단'] != '~')]
df_filtered.drop(0, axis=0, inplace=True)

# 숫자형 변환
df_filtered['발생량'] = pd.to_numeric(df_filtered['발생량'], errors='coerce')


# 시군구별 그룹화
traffic_by_region = df_filtered.groupby(['시도', '시군구', '년'], as_index=False)['발생량'].sum()
```

```python
traffic_by_region['시도'].value_counts()
```

```python
traffic_by_region['시도'] = traffic_by_region['시도'].replace({
    '강원특별자치도': '강원도',
    '전북특별자치도': '전라북도'
})
```

```python
traffic_by_region['지역'] = traffic_by_region['시도'] + ' ' + traffic_by_region['시군구']
```

```python
traffic_by_region.drop(['시도','시군구','년'],axis=1, inplace=True)
```

```python
code=pd.read_excel('/content/code.xlsx')
```

```python
code.columns
```

```python
code['지역 코드'] = code['시/도 코드']*1000+ code['시/군/구 코드']
```

```python
code['지역'] = code['시/도명'] + ' ' + code['시/군/구명']
```

```python
code.drop(['시/도 코드', '시/도명', '시/군/구 코드', '시/군/구명'],axis=1, inplace=True)
```

```python
code
```

```python
# '지역' 컬럼 기준으로 병합
result= traffic_by_region.merge(code, on='지역', how='left')

# 결과 확인
print(result.head())
```

```python
result.drop(['지역'], axis=1, inplace=True)
```

```python
result.to_csv('대중교통 발생량_전처리.csv')
```
